#### 画像中のコーナー検出する

・Harrisコーナー検出:エッジがあらゆる方向に高い木戸変化を持つ領域であると想定し、検出する。\
・Shi-Tomasiコーナー検出:追跡するのに良い特徴が何か。最も強度の高い$N$個のコーナーを見つけ、先のものより良いものが得ることができる。

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
plt.imshow(corners, cmap='gray')
```

#### Scale-Invariant Feature Transform(SIFT)を使う

SIFT:画像の縮尺が変わってもコーナー検出をうまく行かせる。

```python
sift = cv2.SIFT_create()
```

このアルゴリズムは通常、二つのステップで動作する。\
・検出(Detect):画像内の特徴点を特定\
・計算(Compute):各特徴点の実際の特徴量を計算

```python
kp = sift.detect(img_bgr)
img_kp = np.zeros_like(img_bgr)
img_kp = cv2.drawKeypoints(img_rgb, kp, img_kp, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_kp)
```

上図のように画像状のキーポイントの場所にキーポイントの重要度に対応した大きさの円とキーポイントの向きを示す線を描く。

```python
kp, des = sift.compute(img_bgr, kp)
```

```python
des.shape
```

```python
kp2, des2 = sift.detectAndCompute(img_bgr, None)
```

desの値の全ての値がdes2の値とほぼ同じとなる

```python
np.allclose(des, des2)
```
